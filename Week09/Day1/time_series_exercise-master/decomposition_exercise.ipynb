```python
# import pandas
import pandas as pd

# import statsmodels.api as sm
import statsmodels.api as sm

# import matplotlib
import matplotlib.pyplot as plt
```

```python
# load data
data = sm.datasets.co2.load_pandas()
co2 = data.data
```

* print the top 5 rows

```python
co2.head()
```

* Resample the time series using monthly averages and store the new time series as a Series `y`.

```python
# This line converts a timeseries vector into a monthly average table per year
y = co2.resample('M').apply(lambda x: x.max() - x.min()).resample('5AS').apply(lambda x: x.groupby(x.index.month).mean())
```
